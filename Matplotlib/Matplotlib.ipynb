```python
import numpy as np
import matplotlib.pyplot as plt # Useful for plotting and data visualization
```

```python
X = np.linspace(0,10,1000)
Y = np.sin(X)
plt.plot(X,Y) # line plot
```

```python
plt.scatter(X[::10],Y[::10])  #for scatter plot
```

```python
plt.scatter(X[::10],Y[::10],color='green') #change the color
```

```python
plt.plot(X,Y,color='b') # it can be called more than one time with different colors in the plot
plt.plot(X,np.cos(X),color='g') 
```

```python
plt.plot(X,Y +2, '--r') #dashed cyan
plt.plot(X,Y+3, '-.k') #dashot black
plt.plot(X,Y+5, ':g') #dotted green
```
